# Mathematical opr

```python
#Scaler
tens1=torch.tensor([[1,2,3],[4,5,6]])
tens1
```

```python
#these operations will create a new tensor to store result(this is not inplace method)
#add
add=tens1+5

#sub
sub=tens1-2

#mul
mul=tens1*2

#div
div=tens1/2

#exp
exp=tens1**2

#mod
mod=tens1%2
#floor div
floor_div=tens1//2
print(add,sub,mul,div,exp,mod,floor_div)
```

```python
#Element wise opr
a=torch.tensor([[1,2,3],[4,5,6]])
b=torch.tensor([[4,8,2],[7,8,2.2]])
print("add=",a+b,"sub=",a-b,"mul=",a*b,"div=", a/b, "power=", a**b,"mod=" ,a%b,"Floor=" ,a//b)
```

```python
c=torch.tensor([[1.2,-2.5,-1.1],[-3.2,-6.4,9.9]])
print(torch.abs(c), torch.neg(c),torch.max(c), torch.min(c))
```

```python
d=torch.tensor([[1.4,5.6,2.2],[5.8,5.9,1.1]])
d.ceil(), d.floor(), d.trunc(), d.frac(), d.round(), d.sqrt(), d.rsqrt(), d.reciprocal(),d.clamp(2,4)#2 is min and 4 is max..clamping will be done between these
```

# Reduction operations

```python
d=torch.tensor([[1.4,5.6,2.2],[5.8,5.9,1.1]])
print(d)
print("    ")
print(d.sum())#total sum of all elements in tensor
print(d.sum(dim=0))#sum along columns
print(d.sum(dim=1))#sum along rows
```

```python
print(d.mean()) #it only works if dtype of tensor is float
print(d.mean(dim=0),d.mean(dim=1))
print(d.median())
```

```python
print(d.prod(),d.std(), d.var(), d.argmax(), d.argmin()) #argmax gives the index of max no. of tensor and argmin is vice versa
```

# Inplace operations

```python
add1=torch.tensor([[2,3],[6,7]])
add2=torch.tensor([[5,4],[-1,-8]])
print(add1,add2)
```

```python
add1.add_(add2) #this will add add2 tensor to add1 tensor and store resulted or new modified tensor to add1 itself without creating new tensor which will occupy nd waste space
```

```python
add1 #this original add1 is modified so it's an implace method(permanent modification)
```

```python
add2.relu_() #relu converts all negatives to 0
```

# How to copy a tensor

-when we copy a tensor with classical method..the both variables refrences to same memory location instead of creating a new memory..because of this, when we change val of first variable, this also changes the second(copied variable) so to overcome this, we use clone method of pytorch..

```python
#problem
x=torch.tensor([[2,5],[7,8]])
y=x
y
```

```python
x
```

```python
y
```

```python
x.add_(2)
print(x,y) #changing x changed y
```

```python
#Solution
cloned=x.clone()
```

```python
cloned #new copied tensor formed on new memory location
```

```python
x.add_(2) #modifying x tensor
```

```python
cloned #no change in cloned tensor even after changing a tensor
```
