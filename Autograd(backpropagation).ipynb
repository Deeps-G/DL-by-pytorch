# Autograd on single (scaler) perceptron

```python
x=torch.tensor(6.7)
y=torch.tensor(0.0)
```

```python
w=torch.tensor(1.0,requires_grad=True) #grad true because we hv to find derivatives wrt to w and b
b=torch.tensor(0.0,requires_grad=True)
```

```python
w
```

```python
b
```

```python
z=w*x+b
```

```python
z
```

```python
y_pred=torch.sigmoid(z)
```

```python
y_pred
```

```python
import torch.nn.functional as F
loss=F.binary_cross_entropy(y_pred,y)
loss
```

```python
loss.backward()
```

```python
print("derivative or gradient of L wrt w is: " , w.grad , "and L wrt to b is: ", b.grad)
```

#Clearing grad

```python
#when we run passes(forward backward) then gradients acumulates(grad result gets added to previous result and so on)
#so for overcoming this issue as this can be very bad issue as becuase of this we can't get correct values of our grads
#we have to use one func to clear the previous grad's stored result value so that we can re run for next pass without accumulating the result

x=torch.tensor(3.0,requires_grad=True)
```

```python
y=x**2 # forward pass done **this is intermediate node
```

```python
y.backward() #this is first backward pass
x.grad
#if we rerun from y cell and do backward pass and find x grad..then the value is changed from 6 to 12 because new result(6) is added to the previous grad result..but we don't want this..so clear the grad
```

```python
#so we call the func .grad.zero_ to clear the grad
x.grad.zero_()
#see the output..grad is cleared.. now rerun from y node..it will give correct grad(6) at every run but by running this cell after each epoch
```

# How to disable gradient tracking

```python
#when we are done with training the neural network, we should turn off the grad tracking(or doing backpropagation or calc derivatives) as this takes memory..so to save memory, we disable the grad tracking if not required
#we don't require backward pass while making predictions so then we can disable that
#there are three options to do so-
# use .requires_grad_=false
# use detach()
# use torch.no_grad
```

# By .requires_grad_=false

```python
x=torch.tensor(2.0,requires_grad=True)
```

```python
y=x**3
```

```python
x
```

```python
y
```

```python
y.backward() #error because no grad
```

```python
x.grad
```

```python
#now, as training is done..we dont want to use back propagation now..so turn it off
x.requires_grad_(False)
#now see grad is disabled as it doesnt has requires grad true fn in output
#as we can see rerunning the upper backward pass, is giving error as it does nt has grad fn
```

# By Detach

```python
#using detach
x=torch.tensor(2.0,requires_grad=True)
x
```

```python
z= x.detach()
z #now see it has same value as of x tensor but doesnt have requires grad true fn
```

```python
y=x**2
y #this has grad fn as original x had
```

```python
#but when we make by using new node z
y1=z**2
y1 #see this doesnt have grad fn
```

```python
y.backward() #we can do without error
```

```python
y1.backward() #giving error as y1 doesnt have grad fn to backpropagate
```

# by .no_grad() fn

```python
x=torch.tensor(2.0,requires_grad=True)
x
```

```python
with torch.no_grad():
  y=x**2
```

```python
y #no tracking(no grad fn)
```

```python
y.backward() #giving error as no grad
```
